**Separando** x e y

```python
y = dataset["MEDV"]
x = dataset.drop("MEDV", axis = 1)
```

####Normalizando os valores

Aplicando a normalização para deixar os dados na escala de 0-1

```python
from sklearn.preprocessing import MinMaxScaler
min_max_escaler = MinMaxScaler()
escaled = min_max_escaler.fit_transform(x)
x = pd.DataFrame(escaled)
x
```

###Instanciação dos modelos e Treinamento

Instanciação dos modelos a partir da biblioteca Sklearn

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

arv1 = DecisionTreeRegressor(criterion="squared_error",max_depth=5)
arv2 = DecisionTreeRegressor(criterion="friedman_mse",max_depth=5)
knn1 = KNeighborsRegressor(n_neighbors=5, metric="euclidean")
knn2 = KNeighborsRegressor(n_neighbors=5,metric="cosine")
mlp1 = MLPRegressor(hidden_layer_sizes=(20,10), activation='tanh',max_iter=2000)
mlp2 = MLPRegressor(hidden_layer_sizes=(20,10), activation='logistic',max_iter=2000)
svm1 = SVR(kernel="linear", C=100, gamma="auto")
svm2 = SVR(kernel="poly", gamma="scale", C =100)
forest1 = RandomForestRegressor(n_estimators=200, criterion="squared_error")
forest2 = RandomForestRegressor(n_estimators=200, criterion="friedman_mse")
gradient = GradientBoostingRegressor(n_estimators= 200)
```

```python
models = {}
models["Decision Tree 1"] = [arv1, [], [], []]
models["Decision Tree"] = [arv2, [], [], []]
models["KNN 1"] = [knn1, [], [], []]
models["KNN 2"] = [knn2, [], [], []]
models["Rede Neural MLP 1"] = [mlp1, [], [], []]
models["Rede Neural MLP 2"] = [mlp2, [], [], []]
models["SVM 1"] = [svm1, [], [], []]
models["SVM 2"] = [svm2, [], [], []]
models["Random Forest 1"] = [forest1, [], [], []]
models["Random Forest 2"] = [forest2, [], [], []]
models["Gradiente Boost"] = [gradient, [], [], []]
```

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
result = {}

for i in range(10):
   x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2)
   for name, [model, mse_list, rmse_list, mae_list] in models.items():
      model.fit(x_train, y_train)
      result = model.predict(x_test)
      mse = round(metrics.mean_squared_error(result, y_test), 3)
      rmse = round(metrics.mean_absolute_error(result, y_test), 3)
      mae = round(metrics.mean_squared_error(result, y_test, squared=False), 3)
      mse_list.append(mse)
      rmse_list.append(rmse)
      mae_list.append(mae)



```

```python
list_mse_media = []
list_rmse_media = []
list_mae_media = []

for name, [model, mse_list, rmse_list, mae_list] in models.items():
    print("Modelo: {}".format(name))
    print("Lista de valores de MSE: {}".format(mse_list))
    print("Lista de valores de RMSE: {}".format(rmse_list))
    print("Lista de valores de MAE: {}".format(mae_list))
    media1 = round(sum(mse_list)/len(mse_list), 4)
    media2 = round(sum(rmse_list)/len(rmse_list), 4)
    media3 = round(sum(mae_list)/len(mae_list), 4)
    print("Média MSE: {} ".format(media1))
    print("Média RMSE: {} ".format(media2))
    print("Média MAE: {} ".format(media3))
    list_mse_media.append((name, media1))
    list_rmse_media.append((name, media2))
    list_mae_media.append((name, media3))
    print()
```

```python
import random as rd
import matplotlib.pyplot as plt

colors = ['red', 'green', 'blue', 'orange', 'purple', 'pink', 'brown', 'gray', 'cyan', 'yellow']

for name, [model, mse_list, rmse_list, mae_list] in models.items():
    índice = range(len(mse_list))
    media_mse = round(sum(mse_list) / len(mse_list), 3)
    media_rmse = sum(rmse_list) / len(rmse_list)
    media_mae = sum(mae_list) / len(mae_list)

    # Gráfico para MSE
    plt.bar(índice, mse_list, label='MSE Individual', color=rd.choice(colors))
    plt.axhline(y=media_mse, color='r', linestyle='--', label='Média de MSE')
    plt.xlabel('Execução')
    plt.ylabel('MSE')
    plt.title('Mean Squared Error para o Modelo: ' + name)
    plt.xticks(índice, [str(i + 1) for i in índice])
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Gráfico para RMSE
    plt.bar(índice, rmse_list, label='RMSE Individual', color=rd.choice(colors))
    plt.axhline(y=media_rmse, color='r', linestyle='--', label='Média de RMSE')
    plt.xlabel('Execução')
    plt.ylabel('RMSE')
    plt.title('Root Mean Squared Error para o Modelo: ' + name)
    plt.xticks(índice, [str(i + 1) for i in índice])
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Gráfico para MAE
    plt.bar(índice, mae_list, label='MAE Individual', color=rd.choice(colors))
    plt.axhline(y=media_mae, color='r', linestyle='--', label='Média de MAE')
    plt.xlabel('Execução')
    plt.ylabel('MAE')
    plt.title('Mean Absolute Error para o Modelo: ' + name)
    plt.xticks(índice, [str(i + 1) for i in índice])
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
import random as rd
import matplotlib.pyplot as plt

model_names = []
mse_means = []
rmse_means = []
mae_means = []

for name, [model, mse_list, rmse_list, mae_list] in models.items():
    model_names.append(name)
    mse_mean = round(sum(mse_list) / len(mse_list),3)
    rmse_mean = round(sum(rmse_list) / len(rmse_list),3)
    mae_mean = round(sum(mae_list) / len(mae_list),3)
    mse_means.append(mse_mean)
    rmse_means.append(rmse_mean)
    mae_means.append(mae_mean)



# Gráfico de barras para comparar modelos pelo MSE
colors = ["red", "blue", "green", "orange", "purple", "pink", "brown", "cyan", "magenta", "yellow", "lime", "teal"]
plt.figure(figsize=(18, 6))
plt.bar(model_names, mse_means, color=colors)
plt.xlabel('Modelos')
plt.ylabel('MSE Médio')
plt.title('Comparação de Modelos por MSE')
plt.xticks(rotation=45)
plt.tight_layout()
for i, v in enumerate(mse_means):
    plt.text(i, v, str(v), ha='center', va='bottom')
plt.show()
```

```python
# Gráfico de barras para comparar modelos pelo RMSE
plt.figure(figsize=(18, 6))
plt.bar(model_names, rmse_means, color=colors)
plt.xlabel('Modelos')
plt.ylabel('RMSE Médio')
plt.title('Comparação de Modelos por RMSE')
plt.xticks(rotation=45)
plt.tight_layout()
for i, v in enumerate(rmse_means):
    plt.text(i, v, str(v), ha='center', va='bottom')
plt.show()
```

```python
# Gráfico de barras para comparar modelos pelo MAE
plt.figure(figsize=(18, 6))
plt.bar(model_names, mae_means, color=colors)
plt.xlabel('Modelos')
plt.ylabel('MAE Médio')
plt.title('Comparação de Modelos por MAE')
plt.xticks(rotation=45)
plt.tight_layout()
for i, v in enumerate(mae_means):
    plt.text(i, v, str(v), ha='center', va='bottom')
plt.show()
```
